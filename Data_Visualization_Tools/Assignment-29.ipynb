```python
# Q1. Load the "titanic" dataset using the load_dataset function of seaborn. Use Plotly express to plot a
# scatter plot for age and fare columns in the titanic dataset.
```

```python
import seaborn as sns
import plotly.graph_objects as go
```

```python
titanic=sns.load_dataset('titanic')
```

```python
titanic
```

```python
fig=go.Figure()
fig.add_trace(go.Scatter(x=titanic['age'],y=titanic['fare'],mode='markers'))
```

```python
# Q3. Using the tips dataset in the Plotly library, Plot a histogram for x= "sex" and y="total_bill" column in
# the tips dataset. Also, use the "smoker" column with the pattern_shape parameter and the "day"
# column with the color parameter.
```

```python
import plotly.express as px
df = px.data.tips()
fig = px.histogram(df, x="total_bill", y="tip", color="sex",pattern_shape="smoker")
fig.show()
```

```python
# Q5. What is Distplot? Using Plotly express, plot a distplot.
# Ans: A Distplot or distribution plot, depicts the variation in the data distribution.
# The Distplot shows the statistical dispersion of a parameter about the dispersion relation, which is the unitary model parameter.
# The data histogram distribution for a single column is displayed via the distplot() function.
```

```python
import plotly.figure_factory as ff
import numpy as np

# Add histogram data
x1 = np.random.randn(200) - 2
x2 = np.random.randn(200)
x3 = np.random.randn(200) + 2
x4 = np.random.randn(200) + 4

# Group data together
hist_data = [x1, x2, x3, x4]

group_labels = ['Group 1', 'Group 2', 'Group 3', 'Group 4']

# Create distplot with custom bin_size
fig = ff.create_distplot(hist_data, group_labels, bin_size=.2)
fig.show()
```
